```python
# Python program to replace text in a file
x = 'I-LOC'
y = 'B-LOC'
 
# file.txt should be replaced with
# the actual text file name
with open("deu.testb_copy.txt", "r+", encoding='latin-1') as f:

    # each sentence becomes an element in the list l
    lines = f.readlines()

    # acts as a counter to know the
    # index of the element to be replaced
    c = 0
    prev_label = 'O'

    label_dict = {
        'I-LOC': 'B-LOC',
        'I-MISC': 'B-MISC',
        'I-PER': 'B-PER',
        'I-ORG': 'B-ORG'
    }

    for indx, line in enumerate(lines):
        label = line.split()[-1] if len(line)>1 else ''

        if label in ['I-LOC', 'I-MISC', 'I-PER', 'I-ORG'] and label != prev_label:
            prev_label = label

            lines[indx] = line.replace(label, label_dict[label])

```
